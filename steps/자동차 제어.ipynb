```python
from pop import Pilot
from popAssist import *

Car = Pilot.AutoCar()

def userAction(text):
    action = False
    
    print(text)
    
    if text.find("전진") != -1:
        Car.forward()
        action = True
    elif text.find("후진") != -1:
        Car.backward()
        action = True
    elif text.find("정지") != -1:
        Car.stop()
        action = True
        
    return action
stream = create_conversation_stream()
ga = GAssistant(stream, local_device_handler=userAction)


try:
    def onStart():
        print(">>>Start recording...")
        
    while True:
        ga.assist(onStart)
except KeyboardInterrupt:
    stream.close()
```

```python
import cv2
from pop import Pilot, Util

# 카메라 및 자동차 객체 생성
Util.enable_imshow()

cam_string = Util.gstrmer(width=640, height=480)
camera = cv2.VideoCapture(cam_string, cv2.CAP_GSTREAMER)
if not camera.isOpened():
    print("Not found camera")

car = Pilot.AutoCar()
js = Pilot.Joystick()

try:
    while True:
        ret, frame = camera.read()

        if ret:
            cv2.imshow("Camera Feed", frame)

        if js.isConnect():
            # Joystick Y축: 전진 및 후진
            if js.readJoystickY() < 0:
                car.forward(abs(js.readJoystickY())*100)
            elif js.readJoystickY() > 0:
                car.backward(abs(js.readJoystickY())*100)
            else:
                car.stop()

            # (Optional) Joystick X축: 좌/우 방향 제어
            if js.readJoystickX() < 0:
                car.steering(-1)
            elif js.readJoystickX() > 0:
                car.steering(1)
            else:
                car.steering(0)

        if cv2.waitKey(1) == ord('q'):
            break

finally:
    car.stop()
    camera.release()
    cv2.destroyAllWindows()
```

```python
import cv2
from pop import Util

# GStreamer 파이프라인 설정 확인
cam = Util.gstreamer(width=640, height=480)
print("GStreamer Pipeline:", cam)

# 카메라 초기화
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)

# 카메라 열기 확인
if not camera.isOpened():
    print("Failed to open camera!")
else:
    print("Camera opened successfully.")

# 비디오 쓰기 설정
fourcc = cv2.VideoWriter_fourcc(*"X264")
out = cv2.VideoWriter("output.avi", fourcc, 30, (640,480))

# 프레임 캡처 및 저장
for _ in range(120):
    ret, frame = camera.read()
    if not ret:
        print("Failed to grab frame")
        break

    # 프레임 처리 (여기서는 그레이스케일 변환)
    frameGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 비디오 파일에 쓰기
    out.write(frameGray)
    cv2.imshow("Frame", frameGray)

# 자원 해제
camera.release()
out.release()
cv2.destroyAllWindows()
```
